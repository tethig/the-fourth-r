# Heatmap Tutorial

In this tutorial we will explore using ggplot2 to make a heatmap from some 2D data.

## Requirements

```r
library(tidyverse) # this includes ggplot2
library(readxl) # for reading Excel file
```

Home-made rescale function (equivalent to the rescale() from scales package):

```r
rescale <- function(vec){
    floor = min(vec)
    extent = max(vec) - floor
    return( as.vector( scale(vec, center=floor, scale=extent) ) )
}
```

*Note the default behaviour of scale is to subtract the mean (from center arg) and divide by the standard deviation (from scale arg). See ?scale*

## Loading Data

We're going to load the data from an Excel file containing toy microarray data in multiple sheets:

```r
for (i in 1:4) {
  assign( paste("heat",i,sep=""), read_excel("microarrays.xlsx",sheet=i) )
}
```

The assign function created some objects (one per sheet):

```r
ls()
```

Let's look at one (because we used the read_excel function this is technically a *tibble* not a dataframe; see: https://github.com/tidyverse/tibble):

```r
heat2
```

## Heatmap with R Base

We'll start by making a heatmap using the base package. We need to do some work to convert the tibble (or data frame) into a suitable matrix. Our first problem is that column 1 should be taken out and used to give the row names of a numeric matrix:

```r
harry <- data.matrix(heat2[,2:9])
dimnames(harry) <- list(heat2[[1]], colnames(heat2)[2:9])
```

Note that dimnames takes a list with two items in: row names then column names. The [[1]] notation selects the first column, which we give as row names and the column names must be truncated between we are cutting out the first column. Now, let's take a look at harry:

```r
harry
```

That's better, but for the graph we are about to draw we need to transpose this matrix (rotate so the columns become rows and vice versa):

```r
harry <- t(harry) # t stands for transpose
harry
```

That was complex. Now we're in a position to create a heatmap. You might notice a few awkward bits. First, the matrix function plots the row names along the x axis (that's why we had to transpose). Second, the heatmap colours get redder as they decrease so we had to use the rev() to reverse them:

```r
op <- par(cex.axis=0.6, las=2)
image(x=1:8, y=1:6, z=harry, col=rev(heat.colors(12)), xlab='', ylab='', axes=F)
axis(1, at=1:8, labels=rownames(harry))
axis(2, at=1:6, labels=colnames(harry))
title(xlab="sample", ylab="probe", cex.main=2)
par(op)
```

So, there's quite a few awkward problems here and, in general, the colour scale is not very sensitive. Let's go back to the tibbles/data frames and adjust them for analysis by ggplot2. Oh, yes, and let's clean up:

```r
rm(harry,op)
```

## Gathering the Data

Now we will use the gather function to bring each observation into a separate row and specify the patient as a variable. We also add a column with a unique value to each dataframe and place this as the first column:

```r
heat1 <- heat1 %>%
gather(pat1:pat8, key="patient", value="intensity") %>%
mutate(exp="Exp1") %>%
select(exp, everything())

heat2 <- heat2 %>%
gather(pat1:pat8, key="patient", value="intensity") %>%
mutate(exp="Exp2") %>%
select(exp, everything())

heat3 <- heat3 %>%
gather(pat1:pat8, key="patient", value="intensity") %>%
mutate(exp="Exp3") %>%
select(exp, everything())

heat4 <- heat4 %>%
gather(pat1:pat8, key="patient", value="intensity") %>%
mutate(exp="Exp4") %>%
select(exp, everything())
```

Let's concatenate these tibbles into a new dataframe, rename one column and preview:

```r
heatall <- bind_rows(heat1,heat2,heat3,heat4) %>%
rename(probe = X__1)
head(heatall)
```

Let's make sure that factors are properly identified and summarize the new dataframe:

```r
heatall$exp <- factor(heatall$exp)
heatall$probe <- factor(heatall$probe)
heatall$patient <- factor(heatall$patient)
summary(heatall)
```

## Rescaling the Data

We are going to rescale the data, meaning simply that we change all the values so that they lie between 0 and 1. It's important to start by thinking about what the purpose of scaling your data is. You may wish simply to do so for the colour scale. Or you may wish to do so to equalize the spread across different subsets of the data (e.g., patients). 

### Global Rescaling

We are using the rescale function I created at the beginning and applying it to the entire intensity vector:

```r
heatall$rescale <- rescale(heatall$intensity)
head(heatall)
```

### Grouped Rescaling

Alternatively, you may prefer to scale on subsets of the data. In this version, we are scaling within each patient:

```r
patall <- heatall %>%
group_by(patient) %>%
mutate(rescale = rescale(intensity))
head(patall)
#ddply(heatall, .(patient), transform, rescale = rescale(intensity)) # if you're still using plyr
```

Let's tidy up:

```r
rm(heat1, heat2, heat3, heat4, i)
ls()
```

## Graphing the Data

### Heatmap with Panels

Let's draw a heatmap using the ggplot() function and using facet_grid to display each experiment separately:

```r
basic <- ggplot(heatall, aes(x=patient, y=probe))
basic <- basic + geom_tile(aes(fill = rescale), colour = "white")
basic <- basic + scale_fill_gradient2(low = "white", high = "red") # for B&W try: low = "white", high = "black", na.value = "darkred"
basic <- basic + theme(axis.text.x = element_text(angle = 90, hjust = 1))
basic <- basic + facet_wrap(~exp) # or facet_grid(.~exp)
basic
```

Note that if you have multiple factor variables facet_grid can use rows and columns for display. See here:
http://ggplot2.tidyverse.org/reference/facet_grid.html. Let's save this plot:

```r
ggsave("microarray_wrapmap.pdf", width=8, height=6, units="in") # do 12 by 3 if facet_grid
```

If you ran the above with the "patall" dataframe you could take a look at the data when it is rescaled within each patient...

### Making Multiple Plots

We can also make multiple figures by manually subsetting the data:

```r
plots <- list()
for (i in 1:4) {
    e_str = paste("Exp", i, sep="") ;
    plt <- heatall %>%
    filter(exp == e_str) %>%
    ggplot(mapping = aes(x=patient, y=probe)) +
    geom_tile(aes(fill = rescale), colour = "white") +
    scale_fill_gradient2(low = "white", high = "red") +
    ggtitle(paste("Experiment", i))
  plots[[i]] <- plt
}
```

Let's go ahead and save these:

```r
pdf("microarray_multimaps.pdf", width=6, height=4)
plots
dev.off()
```

We can pass this list to a multiple plotting function such as the one provided here: http://www.cookbook-r.com/Graphs/Multiple_graphs_on_one_page_(ggplot2)/.

### Bar Chart with Panels

Let's draw a bar chart with side-by-side columns and facetting - using this we can see the raw values of intensity:

```r
ggplot(heatall, aes(x=patient, y=intensity, fill=probe)) +
  geom_bar(position="dodge", stat="identity", colour="black") +
  facet_grid(exp~.)
```

Let's save this one too:

```r
ggsave("microarray_barchart.pdf", width=10, height=8, units="in")
```

## How I made these data

If you're wondering how I prepared the data for this example:

```r
rnorm(48, 80, 8) -> f1
rnorm(48, 100, 16) -> f2
rnorm(48, 90, 8) -> f3
rnorm(48, 110, 4) -> f4

probes <- c("probe1", "probe2", "probe3", "probe4", "probe5", "probe6")
patients <- c("pat1", "pat2", "pat3", "pat4", "pat5", "pat6", "pat7", "pat8")

matrix(f1, 6, 8, dimnames=list(probes, patients)) -> mat1
matrix(f2, 6, 8, dimnames=list(probes, patients)) -> mat2
matrix(f3, 6, 8, dimnames=list(probes, patients)) -> mat3
matrix(f4, 6, 8, dimnames=list(probes, patients)) -> mat4

write.csv(mat1, file="mat1.csv", quote=F)
write.csv(mat2, file="mat2.csv", quote=F)
write.csv(mat3, file="mat3.csv", quote=F)
write.csv(mat4, file="mat4.csv", quote=F)
```

These data were then concatenated in Excel.
